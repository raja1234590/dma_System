```python
import time
import numpy as np

SIZE = 8  # Elements 1 to 8

# ---- WITHOUT DMA: CPU copies element by element ----
def cpu_transfer(src, dst):
    for i in range(len(src)):
        dst[i] = src[i]
        print(f"  CPU copying element {i+1}: {src[i]} -> dst[{i+1}]")



src = np.array([1, 2, 3, 4, 5, 6, 7, 8], dtype=np.uint8)

# Without DMA
dst1 = np.zeros(SIZE, dtype=np.uint8)
print("=== WITHOUT DMA (CPU copies one by one) ===")
t0 = time.time()
cpu_transfer(src, dst1)
t1 = time.time()
print(f"Result: {dst1}")
print(f"Time: {t1 - t0:.6f}s\n")

```
